````
def get_neighbors(row, col, grid):
    col_max, row_max = len(grid[0]), len(grid)
    directions = [
        (-1, 0),  # up (N)
        (-1, 1),  # up-right (NE)
        (0, 1),   # right (E)
        (1, 1),   # down-right (SE)
        (1, 0),   # down (S)
        (1, -1),  # down-left (SW)
        (0, -1),  # left (W)
        (-1, -1)  # up-left (NW)
    ]
    return [(row + dr, col + dc) for dr, dc in directions
             if (0<=row+dr<row_max) and (0<=col+dc<col_max) # check if in bounds
    ]
````

````
def is_accessible(row, col, grid):
    ns = get_neighbors(row, col, grid)
    rolls = sum(1 for r, c in ns if grid[r][c] == "@")
    return rolls < 4
````

````
total = 0
for r in range(len(xs)):
    for c in range(len(xs[0])):
        if xs[r][c] != "@" : continue
        if is_accessible(r,c, xs): total += 1 
total
````

````
def solve(data, fn):
    grid = parse_data(data)
    total = 0
    for r in range(len(grid)):
        for c in range(len(grid[0])):
            if grid[r][c] != "@": continue
            if fn(r, c, grid): total += 1
    return total

assert solve(exs, is_accessible) == 13
````

````
def solve_part2(data):
    grid = parse_data(data)
    total_removed = 0
    while True:
        locs = [] # purge locs
        for r in range(len(grid)):
            for c in range(len(grid[0])):
                if grid[r][c] == "@" and is_accessible(r, c, grid): locs.append((r,c))
        if not locs: break # end of task
        total_removed += len(locs)
        for r, c in locs: grid[r][c] = "."
    return total_removed

assert solve_part2(exs) == 43
````
