````
def next_even_len(n):
    L = len(str(n))
    if L % 2 == 0: return n
    return 10 ** L
````

````
def find_invalids(x):
    lo, hi = map(int, x.split("-"))
    invalids = []
    start = next_even_len(lo)
    k = len(str(start)) // 2
    patt = int(str(start)[:k])
    while True:
        invalid = patt * (10**k + 1)
        if invalid > hi: break
        if invalid >= lo: invalids.append(invalid)
        patt += 1
    return invalids
````

````
# lets test with several examples
for x in xs:
    print(x, find_invalids(x))
````

````
# expand (extend in python) the list for each range 
def part1(xs=xs): 
    invs = []
    for x in xs: invs.extend(find_invalids(x))
    return sum(invs)
p1 = part1(xs)
print(p1)
puzzle.examples[0].answer_a
````
