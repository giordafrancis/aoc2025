````
from typing import NamedTuple
from math import prod

class Problem(NamedTuple):
    nums: list[int]
    op: str
    
    def solve(self): return sum(self.nums) if self.op == '+' else prod(self.nums)
    
    @staticmethod
    def parse_data(data):
        xs = [x.strip().split() for x in data.splitlines()]
        T = list(zip(*xs))
        return [Problem([int(x) for x in row[:-1]], row[-1]) for row in T]
````

````
xs = Problem.parse_data(exs)
xs
````

````
# lets test 
[x.solve() for x in xs]
````

````
def solve(data):
    items = Problem.parse_data(data)
    return sum(x.solve() for x in items)

solve(puzzle.input_data)
````

There are a few good ways to pad strings in Python:

**1. Using `str.zfill(width)` - pads with zeros on the left:**
```python
"42".zfill(5)  # '00042'
```

**2. Using `str.rjust(width, fillchar)` - pads on the left:**
```python
"42".rjust(5, '0')  # '00042'
"42".rjust(5, ' ')  # '   42'
```

**3. Using `str.ljust(width, fillchar)` - pads on the right:**
```python
"42".ljust(5, '0')  # '42000'
```

````
def parse_p2(data):
    rows = data.splitlines()
    ops, nums = rows[-1].replace(' ', ''), rows[:-1]
    out, new_row = [], []
    for i in range(len(nums[0])):
        num = ''.join([row[i] for row in nums]).strip()
        if num == '': 
            out.append(new_row) 
            new_row = []
        else: new_row.append(int(num))
    out.append(new_row) # last item
    return list(zip(out, ops))

parse_p2(exs)
````

````
def solve_p2(data):
    items = parse_p2(data)
    return sum(sum(nums) if op == '+' else prod(nums) for nums, op in items)

solve_p2(exs)
````
