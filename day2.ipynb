Great question! Here are a few known techniques for detecting repeated patterns in a string:

1. **Divisor checking** â€” If a string is made of a pattern repeated n times, then the string length must be divisible by n. So you only need to check potential pattern lengths that are divisors of the total length.

2. **String doubling trick** â€” A clever one: if you concatenate a string with itself and then search for the original string in the *middle* (skipping the first and last character), finding it means the string has a repeating structure.

3. **Iterative halving** â€” Check if it's 2 repeats, then check if each half is itself 2 repeats, and so on. This is similar to your recursion idea but structured.

## Clean code

````
def parse_data(exs:str=exs) -> list[str]:
    "Parse comma-separated ranges into a list"
    return exs.replace('\n', '').split(',')

def ids(x:str) -> range:
    "Generate range of IDs from string format 'start-end'"
    f,l = x.split('-')
    return range(int(f), int(l)+1)

def repeats(id:int) -> bool:
    "Check if ID has repeating pattern using double-and-chop trick"
    s = str(id)
    return s in (s + s)[1:-1]

def count_repeats(ex:str, repeat_fn=repeats) -> int:
    "Sum all repeating IDs in a single range, using a custom repeat-checking function"
    return sum(id for id in ids(ex) if repeat_fn(id))

def solve(examples:list[str], repeat_fn=repeats) -> int:
    "Sum repeating IDs across all ranges, with pluggable repeat-checking"
    return sum(count_repeats(ex, repeat_fn=repeat_fn) for ex in examples)

def repeats_part1(id:int) -> bool:
    "Check if ID splits into two identical halves"
    s = str(id)
    length = len(s)
    if length % 2 != 0: return False
    return s[0:length//2] == s[length//2:]

def repeats_part2(id:int) -> bool:
    "Check if ID has repeating pattern using double-and-chop trick"
    s = str(id)
    return s in (s + s)[1:-1]
````

Great idea! Here's an Anki card design for the string doubling trick:

---

**Front:**

How can you detect if a string contains a repeating pattern (like "abcabc" or "121212") using concatenation?

---

**Back:**

**The String Doubling Trick:**

```python
s in (s + s)[1:-1]
```

If this is `True`, then `s` has a repeating pattern.

**Why it works:**

When you double a repeating string and chop off the first and last character, the pattern "realigns" in the middle.

**Example with repeating pattern:**
```
s = "abcabc"
s + s = "abcabcabcabc"
[1:-1] = "bcabcabcab"
         ^------^  (original s found here!)
```

**Example without repeating pattern:**
```
s = "abcdef"
s + s = "abcdefabcdef"
[1:-1] = "bcdefabcde"
         (original s NOT found - ends are broken)
```
