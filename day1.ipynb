Great question! The behavior of the modulo operator with negative numbers is actually *language-specific*, not purely mathematical.

In Python, `(-11) % 100 = 89` because Python's modulo always returns a non-negative result with the same sign as the divisor (100 in this case). It's defined so that: `a == (a // b) * b + (a % b)`.

So for `-11 % 100`:
- `-11 // 100 = -1` (floor division)
- `-11 % 100 = 89` (because `-1 * 100 + 89 = -11`)

This is why your "going left" logic works! When you do `(position - step) % 100`, if the subtraction goes negative, Python automatically wraps it around to the correct positive value (0-99 range).

**Why the asymmetry?** Going left can produce negative intermediate values (which wrap around), while going right produces positive values that might exceed 99 (also wrapping). The wrap-around at 0 behaves differently in each direction because of how the subtraction vs addition interacts with the modulo.
