````
def max_joltage_12(s):
    result = []
    start = 0  # where we can start looking
    for i in range(12):  # we need 12 digits
        # How far can we look -> We need 12 - i digits total remaining
        end = len(s) - 12 + i + 1  # i was off by one, need to include end of window
        # Find the max digit in the valid range
        window = s[start:end]
        mx_val = max(window)
        mx_idx = window.index(mx_val) + start # keep in bounds of original s
        result.append(mx_val)
        start = mx_idx + 1  # next search starts after this digit
    
    return int(''.join(result))
````

Exactly right! You've got it. 

By picking the **first** occurrence (leftmost) of the maximum digit, you're leaving more digits available for future picks. If you picked the second '9' instead, you'd "waste" the first '9' and have fewer options remaining.

This greedy strategy works because:
1. You always want the largest digit possible at each position
2. Among equal largest digits, picking the leftmost one maximizes your remaining choices
