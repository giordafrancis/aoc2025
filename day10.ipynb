`````
def check_combo(combo, target):
    "Check if button combo produces target state"
    presses = Counter(L(combo).concat())
    togg = {k for k,v in presses.items() if (v % 2) != 0} 
    return togg == target

def min_presses(buttons, state):
    "Find minimum button presses to reach target state"
    target = {i for i,c in enumerate(state) if c == '#'}
    for k in range(len(buttons) + 1):
        for combo in combinations(buttons, k):
            if check_combo(combo, target): return k

min_presses(bs, s)
`````

`````
def parse_data(data):
    "Parse data into list of machines"
    return [parse_machine(line) for line in data.splitlines()]

def part1(data):
    "Find minimum button presses to reach target state for each machine"
    ms = parse_data(data)
    for state, buttons, _ in ms: yield min_presses(buttons, state)
`````

This is called **Integer Linear Programming (ILP)** â€” or specifically **Linear Programming (LP)** when you allow non-integer solutions.

Since you're minimizing a linear objective (total presses) subject to linear constraints (the equations), and requiring non-negative integers, it's an ILP problem.

Python has several tools for this. `scipy.optimize.linprog` handles LP but not integer constraints directly. For ILP, the `pulp` or `scipy.optimize.milp` (Mixed Integer Linear Programming) can work. Another approach some AoC solvers use is `z3`, a powerful constraint solver.
